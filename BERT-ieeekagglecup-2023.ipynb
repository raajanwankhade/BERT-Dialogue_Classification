```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
import re

def clean_text(text):
    #converting text to lowercase
    text = text.lower()
    #removing special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    #removing extra whitespaces
    text = re.sub(r'\s+', ' ', text)
    #removing leading and trailing whitespaces
    text = text.strip()
    
    return text
```

```python
df['line'] = df['line'].apply(clean_text)
```

```python
df
```

```python
def convert_speaker(speaker):
    if 'Jim' in speaker and 'Dwight' not in speaker:
        return 1
    elif 'Dwight' in speaker and 'Jim' not in speaker:
        return 0
    else:
        return speaker

df['speaker'] = df['speaker'].apply(convert_speaker)
df
```

```python
#deleting rows without Jim or Dwight in the 'speaker' column
df = df[df['speaker'].isin([0,1])]
#resetting the index of the DataFrame
df = df.reset_index(drop=True)
```

```python
df
```

```python
df = pd.DataFrame({
    'line': df['line'],
    'speaker': df['speaker']
})
df
```

```python
df.to_csv('cleaned_train.csv',index=False)
```

```python
from sklearn.model_selection import train_test_split

train_data, val_data = train_test_split(df, test_size=0.2, random_state=42)
val_data, test_data = train_test_split(val_data, test_size=0.5, random_state=42)
```

```python
import tensorflow as tf
!pip install transformers
from transformers import BertTokenizer, TFBertForSequenceClassification
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def tokenize_data(line, speaker):
    line = [str(l) for l in line]  # Convert line to a list of strings
    tokenized_inputs = tokenizer(line, padding=True, truncation=True, return_tensors='tf')
    input_ids = tokenized_inputs['input_ids']
    attention_masks = tokenized_inputs['attention_mask']
    return input_ids, attention_masks, speaker
```

```python
train_line, train_speaker = train_data['line'].tolist(), train_data['speaker'].tolist()
val_line, val_speaker = val_data['line'].tolist(), val_data['speaker'].tolist()
test_line, test_speaker = test_data['line'].tolist(), test_data['speaker'].tolist()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train_speaker = label_encoder.fit_transform(train_speaker)
val_speaker = label_encoder.transform(val_speaker)
test_speaker = label_encoder.transform(test_speaker)
```

```python
train_input_ids, train_attention_masks, train_speaker = tokenize_data(train_line, train_speaker)
val_input_ids, val_attention_masks, val_speaker = tokenize_data(val_line, val_speaker)
test_input_ids, test_attention_masks, test_speaker = tokenize_data(test_line, test_speaker)
```

```python
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metric = tf.keras.metrics.SparseCategoricalAccuracy('accuracy')
```

```python
model.compile(optimizer=optimizer, loss=loss, metrics=[metric])
history = model.fit(
    x=(train_input_ids, train_attention_masks),
    y=train_speaker,
    validation_data=((val_input_ids, val_attention_masks), val_speaker),
    batch_size=32,
    epochs=5
)
```

```python
from sklearn.metrics import classification_report
predictions = model.predict((test_input_ids, test_attention_masks))
predicted_labels = predictions.logits.argmax(axis=1)
predicted_labels = label_encoder.inverse_transform(predicted_labels)

true_labels = label_encoder.inverse_transform(test_speaker)

print(classification_report(true_labels, predicted_labels))
```
